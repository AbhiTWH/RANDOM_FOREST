```python
import pandas as pd
```

```python
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

PLT.GRAY():  THIS SETS THE COLORMAP FOR ALL SUBSEQUENT IMAGES TO GRAYSCALE.

FOR I IN RANGE(4):  THIS STARTS A FOR LOOP THAT ITERATES OVER THE FIRST 4 IMAGES IN THE DIGITS DATASET.
                    THE DIGITS DATASET IS A COLLECTION OF IMAGES OF HANDWRITTEN DIGITS (0-9), WHERE EACH IMAGE IS A 2D ARRAY OF PIXEL VALUES.)

PLT.MATSHOW(DIGITS.IMAGES[I]):  WHAT IT DOES: THIS USES MATPLOTLIB'S MATSHOW FUNCTION TO DISPLAY A MATRIX AS AN IMAGE. 
THE ARGUMENT PASSED TO MATSHOW() IS A 2D ARRAY (DIGITS.IMAGES[I]), WHICH REPRESENTS A DIGIT IMAGE FROM THE DATASET.

```python
plt.hot()
for i in range(4):
    plt.matshow(digits.images[i])
    plt.show()
```

```python
digits.data[:5]
```

```python
df=pd.DataFrame(digits.data)
df.head()
```

```python
df['target']=digits.target
```

```python
df[0:12]
```

```python
d=df.drop('target',axis='columns')
y=df.target
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(d,y,test_size=0.2)
```

```python
len(xtrain)
```

```python
len(xtest)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=20)
```

```python
model.fit(xtrain,ytrain)
```

 MODEL.FIT() IS THE METHOD USED TO TRAIN A MACHINE LEARNING MODEL.
THE FIT METHOD TAKES IN THE TRAINING DATA AND TRAINING LABELS
(ALSO CALLED TARGETS OR RESPONSES) AND THEN ADJUSTS THE MODEL'S INTERNAL PARAMETERS TO MAKE PREDICTIONS BASED ON THE PROVIDED DATA.

```python
model.score(d,y)
```

```python
y_predicted=model.predict(xtest)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest, y_predicted)
cm
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('Truth')
plt.show()
```

```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target']=iris.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(df.drop(['target'],axis=1),iris.target,test_size=0.02)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(Xtrain,Ytrain)
```

```python
model.score(Xtest,Ytest)
```

```python
model=RandomForestClassifier(n_estimators=40)
model.fit(Xtrain,Ytrain)
model.score(Xtest,Ytest)
```
